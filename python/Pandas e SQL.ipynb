# Group by

```python
df.describe()
```

```python
summary_partido = df.groupby( ["sigla_partido", "descricao_cargo"] )["total_votos_brasil"].describe()
```

```python
df.groupby( ["sigla_partido",
             "descricao_cargo"] ).agg( {"total_votos_brasil": ["sum", "mean"]} )
```

```python
resultado_partido = df.groupby("sigla_partido")["total_votos_brasil"].sum()

resultado_partido.sort_values(ascending=False, inplace=True)

resultado_partido = resultado_partido.reset_index()

resultado_partido.rename( columns={"total_votos_brasil":"votos_partido"},
                          inplace=True )
```

```python
resultado_partido
```

## Joins

```python
df_full = pd.merge( left=df,
                    right=resultado_partido,
                      how="left",
                      on="sigla_partido" )
```

```python
df_full["share_candidato"] = df_full["total_votos_brasil"] / df_full["votos_partido"]
```

## SQL

```python
from sqlalchemy import create_engine
```

```python
con = create_engine( "sqlite:////home/teo/Documentos/conre/conre-2/data/my_database.db" )
```

```python
con.table_names()
```

```python
query = '''

SELECT ano_eleicao,
        sum(valor) as valor_total

FROM bens_candidatos
GROUP BY ano_eleicao

'''
```

```python
df_sql = pd.read_sql_query( query, con )
```

```python
df_sql = df_sql[ df_sql["ano_eleicao"] != "ano_eleicao" ].copy()
```

```python
df_sql["ano_eleicao"] = df_sql["ano_eleicao"].astype(int)
```

```python
df_sql.to_sql("ano_patrimonio", con, index=False)
```
